# STANDARDIZATION

```python
import pandas as pd
```

```python
df = pd.read_csv('500hits.csv', encoding='latin-1')
```

```python
df.head()
```

```python
df = df.drop(columns= ['PLAYER', 'CS'])
```

```python
df.info()
```

```python
df.describe().round(3)
```

```python
X1 = df.iloc[:, 0:13]
```

```python
X2 = df.iloc[:, 0:13]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaleStandard = StandardScaler()
```

```python
X1 = scaleStandard.fit_transform(X1)
```

```python
X1.shape
```

```python
X1 = pd.DataFrame(X1, columns=[
    'YRS','G','AB','R','H','2B','3B',
    'HR','RBI','BB','SO','SB','BA'
])
```

```python
X1.head()
```

```python
X1.describe().round(3)
```

# NORMALIZATION

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaleMinMax = MinMaxScaler(feature_range=(0,1))
```

```python
X2 = scaleMinMax.fit_transform(X2)
```

```python
X2 = pd.DataFrame(X2, columns=[
    'YRS','G','AB','R','H','2B','3B',
    'HR','RBI','BB','SO','SB','BA'
])
```

```python
X2.head()
```

```python
X2.describe().round(3)
```
