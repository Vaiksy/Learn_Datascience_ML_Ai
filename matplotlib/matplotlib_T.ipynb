```python
x = ['Mon', 'Tues', 'Wed', 'Thur', 'Fri'] # x-axis
y = [10, 15, 7, 20, 12] # y-axis
plt.plot(x, y)
plt.title('Bakery sales this week !')
plt.xlabel('Day of the week')
plt.ylabel('Sales per day')
plt.show()
```

# Plot function 

# syntex
* plt.plot(x, y) color='color name', linestyle='line_style', linewidth=value, marker='marker symbol', label='label name'

```python
months = [1, 2, 3, 4]
sales = [1000, 1500, 1200, 1800]
plt.plot(months, sales, color='blue', linestyle='--', linewidth = 2, marker = 'o', label='2025 sales data')
plt.xlabel('months')
plt.ylabel('sales per month')
plt.title('Monthly Sales Data Repot')
plt.legend(loc='upper left', fontsize=12)
plt.grid(color='gray', linestyle=':', linewidth=1)
plt.xlim(1, 4)
plt.ylim(0,2000)
plt.xticks([1,2,3,4], ['M1', 'M2', 'M3', 'M4'])
plt.show()
```

# Bar plot 

```python
# for horizontal chart we have to write 'h'with bar like (plt.barh)
product = ['A', 'B', 'C', 'D']
sales2 = [1000, 1500, 800, 1200]
plt.barh(product, sales2, color='red', label='Sales 2026')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sale Comparision')
plt.legend()
plt.show()
```

```python
# this normal one is for vertical chart
product = ['A', 'B', 'C', 'D']
sales2 = [1000, 1500, 800, 1200]
plt.bar(product, sales2, color='red', label='Sales 2026')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sale Comparision')
plt.legend()
plt.show()
```

# This one is for pie chart

```python
reginos = ['North', 'South', 'East', 'West']
revenue = [3000, 2000, 1500, 1000]
plt.pie(revenue, labels=reginos, autopct='%1.1f%%', colors=['red', 'green', 'blue', 'yellow'])
plt.title('Revenue contribution by region')
plt.show()
```

```python
scores = [12,43,53,57,87,45,49,78,32,11,65,93,71,54,25,84,54]
plt.hist(scores, bins=5, color='purple', edgecolor='black')
plt.xlabel('Score range')
plt.ylabel('Number of student')
plt.title('Score Distribution of student')
plt.show()
```
